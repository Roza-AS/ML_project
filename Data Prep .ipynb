```python
#step 1: Data preperations
''' 
In this notebook we are going to prep our data to do EDA next, 
so we are going to load the dataset, get basic information (how many rows, columns, missing data, data types), 
and do relevant changes, like changing data type object to correct data types like category/datetime.. 
to be able to use these variables later on.
We are going to Remove punctuation from the dataset, narrow categories for optimal use,
extracting year/month/day/hour/minutes/seconds from start_time & end_time, 
then save all these changes to the flatfile.
'''
```

```python
#changing datatype from object to string/datetime/category

category_cols = ['Source', 'Severity', 'Wind_Direction', 'Weather_Condition', 'Sunrise_Sunset',
                'Civil_Twilight', 'Nautical_Twilight', 'Astronomical_Twilight'] # list of columns to convert to category
for col in category_cols:
    df[col] = df[col].astype('category')
    
string_cols = ['ID', 'Description', 'Street', 'City', 'County', 'State', 
               'Zipcode', 'Country', 'Timezone', 'Airport_Code'] # list of columns to convert to string
for col in string_cols:
    df[col] = df[col].astype('string')
    
date_cols = ['Start_Time', 'End_Time', 'Weather_Timestamp']  # list of columns to convert to datetime
for col in date_cols:
    df[col] = pd.to_datetime(df[col], errors='coerce')
```

```python
df.info()
```

```python
df.head()
```

```python
#clean text - removing punctuations

import string
```

```python
# Convert specified columns to string type and create string_df
string_df = df[string_cols].astype(str)

# Remove punctuation in one line
string_df = string_df.map(lambda x: x.translate(str.maketrans('', '', string.punctuation)))

# Assign the cleaned columns back to the original DataFrame
df[string_cols] = string_df

# Display the resulting DataFrame
df
```

```python
#Narrowing Categories
```

```python
#df['Timezone'] = df['Timezone'].astype('category')
#replacing all NA with a new 'unknow' category

df['Timezone'] = df['Timezone'].replace('NA', 'Unknown')
df['Timezone'].value_counts()
```

```python
#checking how many categories are there in wind_direction
df['Wind_Direction'].value_counts()
```

```python
#narrowing the categories in wind_direction to 6 main categories; south/west/east/north/calm/variable
df['Wind_Direction'] = df['Wind_Direction'].replace({'S': 'South', 'W': 'West', 'N': 'North',
                                                    'E': 'East', 'CALM': 'Calm', 'VAR': 'Variable',
                                                    'SSW': 'West', 'WNW': 'West', 'NW': 'West', 'SW': 'West',
                                                    'WSW': 'West', 'SSE': 'East', 'NNW': 'West', 'SE': 'East', 'ESE': 'East',
                                                    'NE': 'East', 'ENE': 'East', 'NNE': 'East', 'calm': 'Calm'})
df['Wind_Direction'].value_counts()
```

```python
#checking how many categories are there in weather_condition
df['Weather_Condition'].value_counts()
```

```python
df['Weather_Condition'].unique()
```

```python
#narrowing the categories in weather_codinition as possible

threshold = 10000
counts = df['Weather_Condition'].value_counts()
common = counts[counts >= threshold].index
df['Weather_Condition'] = df['Weather_Condition'].apply(lambda x: x if x in common else 'Other')
```

```python
df['Weather_Condition'].value_counts() #dropped 144 categories to 25
```

```python
#still could use fine tuning;
df['Weather_Condition'] = df['Weather_Condition'].replace({
    'Mostly Cloudy / Windy': 'Mostly Cloudy',
    'Cloudy / Windy': 'Cloudy',
    'Partly Cloudy / Windy': 'Partly Cloudy',
    'Light Rain with Thunder': 'Light Rain',
    'Thunder in the Vicinity': 'Thunder',
    'Heavy Rain': 'Rain',
    'Light Drizzle': 'Rain'
})
df['Weather_Condition'].value_counts()
```

```python
# extracting year/month/day/hour/minutes/seconds from start_time & end_time

# From start time
df['Start_Year'] = df['Start_Time'].dt.year
df['Start_Month'] = df['Start_Time'].dt.month
df['Start_Day'] = df['Start_Time'].dt.day
df['Start_DayOfWeek'] = df['Start_Time'].dt.dayofweek   # Monday=0
df['Start_Hour'] = df['Start_Time'].dt.hour
df['Start_Minute'] = df['Start_Time'].dt.minute
df['Start_Second'] = df['Start_Time'].dt.second
df['Start_IsWeekend'] = df['Start_DayOfWeek'].isin([5,6]).astype(int) # is_weekend

# From end time 
df['End_Year'] = df['End_Time'].dt.year
df['End_Month'] = df['End_Time'].dt.month
df['End_Day'] = df['End_Time'].dt.day
df['End_DayOfWeek'] = df['End_Time'].dt.dayofweek
df['End_Hour'] = df['End_Time'].dt.hour
df['End_Minute'] = df['End_Time'].dt.minute
df['End_Second'] = df['End_Time'].dt.second
df['End_IsWeekend'] = df['End_DayOfWeek'].isin([5,6]).astype(int)

# duration in minutes --- my target value
df['Duration_Minutes'] = (df['End_Time'] - df['Start_Time']).dt.total_seconds() / 60
# duration in hours
df['Duration_Hours'] = df['Duration_Minutes'] / 60

print(df.head())
```
